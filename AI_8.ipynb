## 8-1 製造業

各国で次世代型製造業への転換を打ち出している。

|国名|指針|備考|
|---|---|---|
|ドイツ|Industrie4.0（インダストリー4.0）|－|
|アメリカ|Industrial Internet|－|
|中国|中国製造2025|－|
|日本|Society5.0|スマート工場の取り組み|

スマート工場の実現のため、工場などの現場にサーバーを分散配置する**エッジコンピューティング**を取り入れ、工場内に設置しているセンサなどから得られる大容量のデータに対して、リアルタイムに高速な処理を可能にする。また、エッジコンピューティングで処理したデータを遠隔地にある**クラウドコンピューティング**に送ることで、クラウド側の処理の負担や遅延を防ぐことが可能となる。

また、ディープラーニングにより、画像認識技術などが発達し、ロボットの眼のような役割ができ、物体の認識が可能となった。そのため、ロボティクスの分野でもロボットの動作を覚えさせるために強化学習が進み、各種センサから取得した**マルチモーダル**な情報に対してディープラーニングを活用することで入出力に設計者の介入を必要としない一気通貫学習ができるようになった。

## 8-2 自動車産業

### 8-2-1 自動運転

米国のSAE(Society of Automotive Engineers)が定めた**SAE J3016**の定義を採用する。

|レベル|概要|監視・対応主体|
|---|---|---|
|レベル0<br>運転自動化なし|運転者がすべての運転タスクを実施|運転者|
|レベル1<br>運転支援|システムが前後、左右のいずれかの車両制御に係る運転タスクのサブタスクを実施|運転者|
|レベル2<br>部分運転自動化|システムが前後、左右の両方の車両制御に係る運転タスクのサブタスクを実施|運転者|
|レベル3<br>条件付運転自動化|・システムがすべての運転タスクを実施(限定領域内)<br>・作動継続が困難な場合の運転者は、システムの介入要求などに対して<br>　適切に応答することを期待される|システム<br>(作動継続が困難な場合運転者)|
|レベル4<br>高度運転自動化|・システムがすべての運転タスクを実施(限定領域内)<br>・作動継続が困難な場合、利用者が応答することは期待されない|システム|
|レベル5<br>完全運転自動化|・システムがすべての運転タスクを実施(限定領域内ではない)<br>・作動継続が困難な場合、利用者が応答することは期待されない|システム|

上記レベルの設定があるが、自動運転の企業のアプローチとして、2つのアプローチがある。

|企業|取り組み|
|---|---|
|自動車メーカー|レベル1から段階的|
|サプライヤー|レベル1から段階的|
|IT企業|レベル3から実現|
|スタートアップ企業|レベル3から実現|

## 8-3 インフラ・農業

### 8-3-1 インフラ

インフラにおけるディープラーニングの活用例として、設備の故障や異常検知を行うメンテナンスの効率化(コンクリートのひび割れ検出、橋梁の損失推定)がある。

センサを取り付けてIoT化しデータを収集、小型無人機（**ドローン**）による空撮を行い、ディープラーニングを用いた解析する事例もある。

ドローンを活用するには定められた飛行ルールに準じなければならず、国から許可が必要な領域は以下である。

1. 空港などの周辺の上空空域
1. 人口集中地区の上空
1. 150m以上の高さの空域

国からの承認が必要な飛行方法は以下である。

1. 夜間飛行
1. 目視外飛行
1. 30m未満の飛行
1. イベント上空飛行
1. 危険物輸送
1. 物件投下

### 8-3-2 農業

日本では農家数が減少し、かつ高齢化が進んでいることもあり、ディープラーニングの活用が求められている現状がある。政府は、**スマート農業**を実現するために取り込みが開始しており、以下の応用に期待している。

1. 熟練者の作業ノウハウのAIによる形式知化
1. 遠隔監視による農機の無人走行システム実現
1. ドローンとセンシング技術やAIの組み合わせによる農薬散布、施肥などの最適化
1. 自動走行農機などの導入・利用に対応した土地改良事業の推進

海外では農地が広大のため、ドローンを活用し、上空から得た写真の画像認識より作物や土壌の状況を分析するサービスもある。
